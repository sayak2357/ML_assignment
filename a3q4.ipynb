```python
import numpy as np
from numpy.linalg import inv
```

```python
def normalEquationRegression(X, Y):
    xt = X.transpose()
    b = np.ones((1,xt.shape[1]))
    xt2 = np.vstack((b,xt))
    x = xt2.transpose()
    print(xt2)
    print(x)
    n1=np.matmul(xt2,x)
    n2=inv(n1)
    n3=np.matmul(n2,xt2)
    n4= np.matmul(n3,Y)
    return n4
    
```

```python
def graient_descent(x,y):
    mcurr=bcurr=0
    n=len(x)
    iterations = 1000
    learningrate=0.001
    for i in range(0,10000):
        print(mcurr,bcurr,i)
        y_predicted= x*mcurr+bcurr
        md= -(2/n)*sum(x*(y-y_predicted))
        bd= -(2/n)*sum(y-y_predicted)
        mcurr=mcurr - md*learningrate
        bcurr=bcurr - bd*learningrate
```

```python
x=np.array([1,3,6])
y=np.array([3,7,13])
```

```python
graient_descent(x,y)
```
