```python

dict1={}
for i in range(1,1001):
    ts2 =int(time.time())
    ts4=ts2*i
    rd= ts4 % 100
    dict1[i]=rd    
bl=[]
bl2=[]
for x in dict1:
    bl.append(dict1[x])
for x in dict1:
    bl2.append(x)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
boys_grades = bl
plt.scatter(boys_grades,bl2, color='g')
plt.xlabel('Grades Range')
plt.ylabel('Grades Scored')
plt.show()
```
