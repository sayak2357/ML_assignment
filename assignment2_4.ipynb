# 4.a

```python
import numpy
```

```python
xd=[]
for i in range(0,50):
    xd.append(i)
```

```python
yd=[]
for i in range(0,50):
    z=numpy.random.randint(1,10 , size=None, dtype='l')
    y= 5*xd[i]+10+z
    yd.append(y)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
indexes = xd
plt.scatter(indexes,yd, color='g')
plt.xlabel('x indices')
plt.ylabel('y =mx +x + noise')
plt.show()
```

# 4.b

```python
import numpy as np
import matplotlib.pyplot as plt
p = np.poly1d(np.polyfit(xd, yd, 5))

t = np.linspace(xd[0],xd[-1], 200)
plt.plot(xd, yd, 'o', t, p(t), '-')
plt.show()
```

# 4.c

```python
bag1=[]
x1=[]
for i in range(0,25):
    z2=numpy.random.randint(1,49 , size=None, dtype='l')
    bag1.append(yd[z2])
    x1.append(z2)
bag2=[]
x2=[]
for i in range(0,25):
    z2=numpy.random.randint(1,49 , size=None, dtype='l')
    bag2.append(yd[z2])
    x2.append(z2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
p = np.poly1d(np.polyfit(x1, bag1, 5))

t = np.linspace(x1[0], x1[-1], 200)
plt.plot(x1, bag1, 'o', t, p(t), '-')
plt.title('BAG 1')
plt.xlabel('x indices')
plt.ylabel('y =mx +x + noise')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
p = np.poly1d(np.polyfit(x2, bag2, 5))

t = np.linspace(x1[0],x1[-1], 200)
plt.plot(x2, bag2, 'o', t, p(t), '-')
plt.title('BAG 2')
plt.xlabel('x indices')
plt.ylabel('y =mx +x + noise')
plt.show()
```
