```python
f=open("iris.csv")
x=[]
y=[]
```

```python
for lines in f:
    print(lines)
    a=lines.split(",")
    print(a[0])
    t1=float(a[0])
    t2=float(a[1])
    t3=float(a[2])
    x.append([t1,t2])
    y.append([t3])
```

```python
import numpy as np
from sklearn.svm import SVC
clf = SVC(gamma='auto')
clf.fit(x,y )
```

```python
print(clf.predict([[-0.8, -1]]))
```
