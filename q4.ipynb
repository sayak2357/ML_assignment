# 4.a) y=mx+c without any noise

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.ensemble import BaggingRegressor
x=[]
for i in range(0,50):
    x.append(i)
y=[]
for i in range(0,50):
    y1= 6*x[i]+10
    y.append(y1)
plt.scatter(x,y, color='y')
plt.xlabel('x')
plt.ylabel('y =mx +c')
plt.show()

```

# added noise (z is a random integer from 1 to 100 for each value of y)

```python
x=[]
for i in range(0,50):
    x.append(i)
y=[]
for i in range(0,50):
    z=numpy.random.randint(1,100,size=None, dtype='l')
    y1=5*x[i]+10+z
    y.append(y1)
plt.scatter(x,y, color='y')
plt.xlabel('x')
plt.ylabel('y =mx +c + noise')
plt.show()
```

# 4.b) 5 degree polyfit for the noisy data generated.

```python
f = np.poly1d(np.polyfit(x, y, 5))

t = np.linspace(0, 50, 200)
plt.plot(x, y, 'o', t, f(t), '-')
plt.show()
```

# 4.c)number of rounds of bagging to 100

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
import copy

import warnings
warnings.filterwarnings("ignore")
#X=data.drop('y', axis=1)
#Y=data['y']
X=np.asarray(x).reshape(-1,1)
Y=np.asarray(y).reshape(-1,1)
x_train, x_test, y_train,y_test=train_test_split(X,Y, test_size=0.2,random_state=100)
model = BaggingRegressor(n_estimators=100, random_state=101)

model.fit(x_train, y_train)
model.score(x_test,y_test)
```
